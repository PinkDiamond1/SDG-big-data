```python
from results import *
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
from matplotlib.transforms import Affine2D
```

```python
sns.set(style='whitegrid')
sns.despine()
```

```python
def extract_data(label, data_dict):
    precision_list = list()
    for iter_nb in range(4):
        precision_list.append(data_dict[iter_nb][label])
    if 4 in data_dict.keys():
        precision_list.append(data_dict[4][label])
    return precision_list
```

```python
output_path = '/home/manuto/Documents/world_bank/bert_twitter_labor/twitter-labor-data/data/fig/precision'
```

```python
labels=['is_hired_1mo', 'is_unemployed', 'job_offer', 'job_search', 'lost_job_1mo']
```

```python
iter_nb = range(5)
```

# top 50

```python
fig,ax = plt.subplots(nrows=1, ncols=5, figsize=(4, 4))
for num, label in enumerate(labels):
    fig,ax = plt.subplots(nrows=1, ncols=5, figsize=(4, 4))
    trans1 = Affine2D().translate(-0.2, 0.0) + ax[num].transData
    trans2 = Affine2D().translate(-0.1, 0.0) + ax[num].transData
    trans3 = Affine2D().translate(+0.1, 0.0) + ax[num].transData
    precision_list_our_method = extract_data(label, our_method_top50_dict)
    precision_list_adaptive = extract_data(label, adaptive_top50_dict)
    precision_list_uncertainty = extract_data(label, uncertainty_top50_dict)
    precision_list_uncertainty_uncalibrated = extract_data(label, uncertainty_uncalibrated_top50_dict)
    ax[num].scatter(iter_nb, precision_list_our_method, c='b', label='explore/exploit retrieval', transform=trans1, marker='o')
    ax[num].scatter(iter_nb, precision_list_adaptive, c='r', label='adaptive retrieval', transform=trans2, marker='v')
    ax[num].scatter(iter_nb, precision_list_uncertainty, c='g', label='uncertainty sampling', marker='s')
    ax[num].scatter(iter_nb, precision_list_uncertainty_uncalibrated, c='m', label='uncertainty sampling (uncalibrated)', transform=trans3, marker='P')
    #ax[num].tick_params(which='both',direction='in',pad=3)
#     ax.locator_params(axis='y',nbins=6)
#     ax[num].set_ylabel('Precision (top50)',fontweight='bold')
#     ax[num].set_xlabel('Iteration number',fontweight='bold')
#     ax[num].set_title(label.replace('_',' ').replace('1mo','').title(),fontweight='bold')
#     ax[num].legend(loc='best',fontsize=9)
#     ax[num].set_ylim([-0.05,1.05])
    sns.despine()
    #plt.savefig(os.path.join(output_path,f'precision_top50_{label}.png'),bbox_inches='tight', format='png' ,dpi=1200, transparent=False)
```

```python
for num, label in enumerate(labels):
    fig,ax = plt.subplots(figsize=(4,4))
    trans1 = Affine2D().translate(-0.2, 0.0) + ax.transData
    trans2 = Affine2D().translate(-0.1, 0.0) + ax.transData
    trans3 = Affine2D().translate(+0.1, 0.0) + ax.transData
    precision_list_our_method = extract_data(label, our_method_top50_dict)
    precision_list_adaptive = extract_data(label, adaptive_top50_dict)
    precision_list_uncertainty = extract_data(label, uncertainty_top50_dict)
    precision_list_uncertainty_uncalibrated = extract_data(label, uncertainty_uncalibrated_top50_dict)
    ax.scatter(iter_nb, precision_list_our_method, c='b', label='explore/exploit retrieval', transform=trans1, marker='o')
    ax.scatter(iter_nb, precision_list_adaptive, c='r', label='adaptive retrieval', transform=trans2, marker='v')
    ax.scatter(iter_nb, precision_list_uncertainty, c='g', label='uncertainty sampling', marker='s')
    ax.scatter(iter_nb, precision_list_uncertainty_uncalibrated, c='m', label='uncertainty sampling (uncalibrated)', transform=trans3, marker='P')
    ax.tick_params(which='both',direction='in',pad=3)
#     ax.locator_params(axis='y',nbins=6)
    ax.set_ylabel('Precision (top50)',fontweight='bold')
    ax.set_xlabel('Iteration number',fontweight='bold')
    ax.set_title(label.replace('_',' ').replace('1mo','').title(),fontweight='bold')
    if num == 4:
        ax.legend(loc='best',fontsize=9)
    ax.set_ylim([-0.05,1.05])
    sns.despine()
    plt.savefig(os.path.join(output_path,f'precision_top50_{label}.png'),bbox_inches='tight', format='png' ,dpi=1200, transparent=False)
```
